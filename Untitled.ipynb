```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 200)
y = np.sin(x) + 0.4*x + 0.4*np.random.randn(200)
```

```python
plt.scatter(x,y)
```

```python
y.max()
```

```python
x = x/10
y = y/y.max()
```

```python
x_train = x.reshape(200,1)
y_train = y
```

$$ f\colon \mathbb{R}^1 \to \mathbb{R}^1 $$

```python
%env KERAS_BACKEND=tensorflow
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
```

```python
model = Sequential()
```

```python
model.add(Dense(50, input_dim=1))
model.add(Activation('sigmoid'))
```

```python
model.add(Dense(50))
model.add(Activation('sigmoid'))
```

```python
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
model.compile(loss='mse', optimizer=SGD(lr=0.01), metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train, y_train, batch_size=20, epochs=20000)
```

```python
Y = model.predict(x_train)
```

```python
plt.scatter(x, y)
plt.plot(x, Y.ravel(), 'r')
```
